```python
# The GroupBy Object
import pandas as pd
```

```python
fortune = pd.read_csv('../../Files/fortune1000.csv', index_col="Rank")
fortune.head()
```

```python
# sums of revenue for each sector
fortune[
    fortune["Sector"] == "Retailing"
]["Revenue"].sum()
```

```python
sectors = fortune.groupby("Sector")
```

```python
len(sectors)
```

```python
sectors.size()
```

```python
sectors\
    .first()

sectors\
    .last()
```

```python
# retrieve a group with the get_group method

fortune.head()
```

```python
sectors
```

```python
sectors.get_group("Energy")
sectors.get_group("Technology")
```

```python
sectors["Revenue"]
```

```python
sectors["Revenue"].sum()
sectors["Employees"].sum()
sectors["Profits"].max()

sectors["Employees"].mean()
sectors["Employees"].min()

sectors[["Revenue", "Profits"]].max()
sectors[["Revenue", "Profits"]].min()
```

```python
# grouping by multiple columns
sectors = fortune.groupby(["Sector", "Industry"])
```

```python
sectors.size()
```

```python
sectors["Revenue"].sum()
```

```python
sectors["Employees"].mean()
```

```python
sectors.agg({
    "Revenue": "sum",
    "Profits": "max"
})
```

```python
# find the two companies in each sector with the most employees
fortune.nlargest(2, "Employees")  # wrong becuase we need to group by sector.
```

```python
sectors.apply(
    lambda sector: sector.nlargest(2, "Employees"),
    include_groups=False
)
```

Merging DataFrames


```python
import pandas as pd

foods = pd.read_csv('../../Files/restaurant_foods.csv')
customers = pd.read_csv('../../Files/restaurant_customers.csv')
week1 = pd.read_csv('../../Files/restaurant_week_1_sales.csv')
week2 = pd.read_csv('../../Files/restaurant_week_2_sales.csv')
```

```python
display(foods)
display(customers)
display(week1)
display(week2)
```

```python
week1.head()
```

```python
week2.head()
```

```python
len(week1)
```

```python
len(week2)
```

```python
# pd.concat([week1,week2], ignore_index=False).sort_index()
pd.concat([week1, week2], keys=["Week 1", "Week 2"]).sort_index()
```

```python
df1 = pd.DataFrame([1, 2, 3, 7], columns=["A"])
df2 = pd.DataFrame([4, 5, 6], columns=["B"])
```

```python
display(df1)
display(df2)
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], axis="index")
```

```python
pd.concat([df1, df2], axis="columns")
```

```python
foods.head()
```

```python
week1.merge(foods, how="left", on="Food ID")
```

```python
week2.head()
```

```python
customers.head()
```

```python
week2.merge(customers, how="left", left_on="Customer ID", right_on="ID")
```

```python
week1.merge(week2, how="left", on="Customer ID",
            suffixes=[" - Week 1", " - Week 2"])
```

```python
week1.merge(week2, how="inner", on=["Customer ID", "Food ID"])
```

```python
condition1 = week1["Customer ID"] == 21
condition2 = week1["Food ID"] == 4
week1[condition1 & condition2]
```

```python
condition1 = week2["Customer ID"] == 21
condition2 = week2["Food ID"] == 4
week2[condition1 & condition2]
```
