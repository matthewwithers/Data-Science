```python
employees = pd.read_csv('../../Files/employees.csv')
employees.info()
```

```python
employees["Start Date"] = pd.to_datetime(
    employees["Start Date"], format="%m/%d/%Y")
```

```python
employees["Last Login Time"] = pd.to_datetime(employees["Last Login Time"], format="%H:%M %p")\
    .dt.time
```

```python
employees["Senior Management"] = employees["Senior Management"].astype(bool)
```

```python
employees.nunique()
```

```python
employees["Gender"] = employees["Gender"].astype("category")
```

```python
employees = pd.read_csv('../../Files/employees.csv',
                        parse_dates=["Start Date"], date_format="%m/%d/%Y")
employees["Last Login Time"] = pd.to_datetime(employees["Last Login Time"], format="%H:%M %p")\
    .dt.time
employees["Senior Management"] = employees["Senior Management"].astype(bool)
employees["Gender"] = employees["Gender"].astype("category")
```
