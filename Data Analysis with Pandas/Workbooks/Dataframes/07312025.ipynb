```python
employees = pd.read_csv('../../Files/employees.csv',
                        parse_dates=["Start Date"], date_format="%m/%d/%Y")
employees["Last Login Time"] = pd.to_datetime(employees["Last Login Time"], format="%H:%M %p")\
    .dt.time
employees["Senior Management"] = employees["Senior Management"].astype(bool)
employees["Gender"] = employees["Gender"].astype("category")
```

```python
employees.dtypes
```

```python
employees[employees["Gender"] == "Female"]
```

```python
# senior_management = employees["Senior Management"] == True
employees[employees["Senior Management"]]  # already stored as a boolean
```

```python
employees[employees["Salary"] > 110000].sort_values(
    ["First Name", "Start Date"], ascending=[True, False])
```

```python
import datetime as dt

employees[employees["Start Date"] >= "2000-01-01"]

employees[employees["Last Login Time"] < dt.time(12, 0, 0)]
```

```python
employees[employees["First Name"] == "Douglas"]
```

```python
display(employees.loc[
    (employees["Gender"] == "Male") &
    (employees["First Name"].isin(["Douglas", "Henry"])) &
    (employees["Team"] == "Legal")
])
```

```python
# female employees who work in marketing
is_female = employees["Gender"] == "Female"
is_in_marketing = employees["Team"] == "Marketing"
employees[is_female & is_in_marketing]

employees.loc[
    (employees["Gender"] == "Female") &
    (employees["Team"] == "Marketing")
][[
    "First Name", "Gender", "Start Date", "Salary"
]]
```
