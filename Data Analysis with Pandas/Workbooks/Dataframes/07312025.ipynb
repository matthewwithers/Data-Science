```python
data_list_of_dicts = [
    {'name': 'Alice', 'age': 30},
    {'name': 'Bob', 'age': 25},
    {'name': 'Bob', 'age': 25},
    {'name': 'Alice', 'age': 30},
]
df = pd.DataFrame(data_list_of_dicts)
display(df)
display(df[~df.duplicated()])
display(df[~df.duplicated(keep="first")])
display(df[~df.duplicated(keep="last")])
display(df[df.duplicated()])
print('drop duplicates')
display(df.drop_duplicates().sort_values(by=""))
```
